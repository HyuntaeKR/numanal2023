# Interpolation

## Lagrange Polynomial Interpolation

```python
import numpy as np
from numpy import typing


def lagrange_interp(data: np.ndarray, x_point: typing.ArrayLike) -> typing.ArrayLike:
    """
    Returns value corresponding to input 'x_point' based on
    Lagrange polynomial interpolation generated by input data.
    """
    num_data_pt = len(data)
    x_data = data[:, 0]
    y_data = data[:, 1]

    def Pfunction(x_point: typing.ArrayLike, idx: int) -> float:
        result_in = 1
        for i in range(num_data_pt):
            if i != idx:
                result_in *= (x_point - x_data[i]) / (x_data[idx] - x_data[i])
        return result_in

    result_out = 0
    for i in range(num_data_pt):
        result_out += y_data[i] * Pfunction(x_point, i)

    return result_out
```

```python
# Example of Lagrange polynomial interpolation
import matplotlib.pyplot as plt

data = np.genfromtxt("./data/ch3_data_for_interpol.txt")
interp = lagrange_interp(data, np.linspace(0, 1, 101))

fig = plt.figure()
plt.scatter(data[:, 0], data[:, 1])
plt.plot(np.linspace(0, 1, 101), interp, "r-")
```
